ZDEFINIOWANIE FUNKCJI WYŚWIETLAJĄCEJ ZDJĘCIA

```python
def show2imgs(im1, im2, title1='Obraz pierwszy', title2='Obraz drugi', size=(10, 10)):
    import matplotlib.pyplot as plt

    f, (ax1, ax2) = plt.subplots(1, 2, figsize=size)
    ax1.imshow(im1, cmap='gray')
    ax1.axis('off')
    ax1.set_title(title1)

    ax2.imshow(im2, cmap='gray')
    ax2.axis('off')
    ax2.set_title(title2)
    plt.show()

   
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

WCZYTANIE ZDJECIA I WYSWIETLENIE GO

```python
im = cv2.imread("zdjecie15.jpg", 0)
th = 150
th, bim = cv2.threshold(im, thresh=th, maxval=255, type=cv2.THRESH_OTSU)

show2imgs(im, bim, title1='Obraz oryginalny', title2='Obraz po binaryzacji', size=(10,10))
```

EROZJA

```python
kernel = np.ones((3,3),np.uint8)

erodeBin = cv2.erode(bim, kernel=kernel, iterations=30)
show2imgs(bim, erodeBin, title1='Obraz oryginalny', title2='Obraz po erozji', size=(10,10))
```

WYLICZENIE ILOŚCI PIKSELI I OBIEKTÓW NA ZDJĘCIU

```python
from scipy import ndimage as ndi
label_objects, nb_labels = ndi.label(erodeBin)

sizes = np.bincount(label_objects.ravel())
mask_sizes = sizes > 20
mask_sizes[0] = 0

figures = mask_sizes[label_objects]

plt.figure(figsize=(10,10))
plt.imshow(figures, cmap="hot")
plt.axis('on')
plt.suptitle('Analizowany obraz: ')
plt.show()

print("Rozmiar obrazu: " + str(figures.shape))
print("Liczba obiektow na obrazie: " + str(nb_labels))
```

FUNKCJA ZAPISUJĄCA PUNKTY KRAWĘDZI OBIEKTÓW DO LISTY

```python
def getFigure(labelledImage, objNumber):
    
    points = []
    for y in range(labelledImage.shape[0]):
        for x in range(labelledImage.shape[1]):
            if labelledImage[y,x] == objNumber:
                points.append((y,x))

    return points    
```

FUNKCJA LICZĄCA ŚRODEK CIĘŻKOŚCI

```python
def cog2(points):
    mx=0
    my=0
    for (y,x) in points:
        mx = mx + x
        my = my + y
    mx = mx/len(points)
    my = my/len(points)
    
    return [my, mx]
```

FUNKCJA LICZĄCA WSPÓŁCZYNNIK Blaira-Blissa 

```python
from scipy.spatial import distance
import math

def computeBB(points):
    
    s = len(points)
    my,mx = cog2(pts)
    
    r = 0
    for point in points:
         r = r + distance.euclidean(point,(my,mx))**2
            
    return s/(math.sqrt(2*math.pi*r))
```

FUNKCJA LICZĄCA WSPÓŁCZYNNIK Fereta

```python
def computeFeret(points):
    
    px = [x for (y,x) in points]
    py = [y for (y,x) in points]
    
    fx = max(px) - min(px)
    fy = max(py) - min(py)
    
    return float(fy)/float(fx)   
```

WYWOŁANIE FUNKCJI I WYPISANIE WYNIKÓW

```python
for i in range(nb_labels):
    pts = getFigure(label_objects, i+1)
    bb = computeBB(pts)
    feret = computeFeret(pts)
    
    print('Liczba punktow: ',len(pts),'\nSrodek ciezkosci: ', cog2(pts),'\nBlair-Bliss: ', bb,'\nFeret: ',feret, '\n---\n')
```
